# else는 필수아님.


```python
score = int(input("점수 입력>> "))
if score <=100 and score>= 90: ## 90<= score <= 100: 이렇게도 된다.
    print(f'{score}은 A학점입니다.')
elif score< 90 and score>=80: ## 90은 없어도 된다. 왜냐면 이미 첫번째 조건에서 false되어서 내려왔기 때문에.
    print(f'{score}은 B학점입니다.')
elif score<80 and score>=70:
    print(f'{score}은 C학점입니다.')
elif score<70 and score>=60:
    print(f'{score}은 D학점입니다.')
else:
    print(f'{score}은 F학점입니다.')
```
